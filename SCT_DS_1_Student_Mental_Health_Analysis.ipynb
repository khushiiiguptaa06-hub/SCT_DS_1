```python
import seaborn as sns

sns.set_style("whitegrid")
sns.set_palette("pastel")
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Choose your gender',
    data=df,
    palette='pastel'
)

plt.title('Gender Distribution')
plt.show()
```

Observation

The chart shows the distribution of male and female students in the survey.

It helps us understand whether the dataset is balanced across genders.

Any imbalance should be considered while interpreting mental health trends.

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Age'],
    bins=10,
    kde=True,
    color='#A8DADC'
)

plt.title('Age Distribution')
plt.show()
```

Observation

Most respondents belong to the typical university age range.

The histogram helps identify the most common age group.

Very few students fall outside the main age cluster.

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Do you have Depression?',
    data=df,
    palette=['#CDB4DB','#FFAFCC']
)

plt.title('Depression Distribution')
plt.xlabel('Depression')
plt.ylabel('Count')
plt.show()
```

Observation

This visualization shows the number of students experiencing depression.

It provides an overview of mental health conditions among respondents.

The results can be used to assess the prevalence of depression within the dataset.

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='Do you have Anxiety?',
    data=df,
    palette=['#FFD6A5','#FDFFB6']
)

plt.title('Anxiety Distribution')
plt.xlabel('Anxiety')
plt.ylabel('Count')
plt.show()
```

Observation

A considerable proportion of students reported experiencing anxiety, indicating that anxiety is a common mental health concern among the surveyed population.

Anxiety appears to be an important mental health concern among students.

Comparing anxiety and depression distributions can reveal broader trends.

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x='Choose your gender',
    hue='Do you have Depression?',
    data=df,
    palette=['#A0C4FF','#FFAFCC']
)

plt.title('Depression by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Depression')
plt.show()
```

Observation

Depression is observed across both genders.

This chart allows us to compare mental health patterns between male and female students.

Further analysis may help explain differences in reported depression levels.

```python
plt.figure(figsize=(10,6))

top_courses = df['What is your course?'].value_counts().head(10)

sns.barplot(
    x=top_courses.values,
    y=top_courses.index,
    palette='pastel'
)

plt.title('Top 10 Courses by Student Count')
plt.xlabel('Number of Students')
plt.ylabel('Course')
plt.show()
```

Observation

The majority of respondents belong to a small number of courses, with Pendidikan Islam, Engineering, and BIT contributing the highest number of participants. Several courses have very few respondents, indicating an uneven distribution across academic disciplines.

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x='Choose your gender',
    hue='Do you have Anxiety?',
    data=df,
    palette=['#A8DADC', '#FFAFCC']
)

plt.title('Anxiety by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Anxiety')
plt.show()
```

Observation

Anxiety is observed among both male and female students.

The chart helps compare anxiety prevalence across genders.

Differences may indicate varying mental health experiences.

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x='Do you have Depression?',
    hue='Do you have Panic attack?',
    data=df,
    palette=['#BDE0FE', '#FFAFCC']
)

plt.title('Depression vs Panic Attack')
plt.xlabel('Depression')
plt.ylabel('Count')
plt.legend(title='Panic Attack')
plt.show()
```

Observation

Students reporting depression appear more likely to experience panic attacks.

The graph suggests a potential relationship between these mental health conditions.

Further analysis could explore this connection in greater detail.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Your current year of Study',
    hue='Do you have Depression?',
    data=df,
    palette=['#A8DADC', '#FFAFCC']
)

plt.title('Depression by Year of Study')
plt.xlabel('Year of Study')
plt.ylabel('Count')
plt.show()
```

Observation

Depression levels vary across academic years.

Certain year groups may experience greater academic pressure.

This visualization highlights how mental health can differ during a student's academic journey.

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    pd.crosstab(
        df['Do you have Depression?'],
        df['Do you have Panic attack?']
    ),
    annot=True,
    cmap='Pastel1'
)

plt.title('Relationship Between Depression and Panic Attack')
plt.show()
```

Observation

The largest group of students (50) reported neither depression nor panic attacks.

Among students experiencing depression, the number reporting panic attacks (17) is nearly equal to those not reporting panic attacks (18).

Students without depression are considerably less likely to experience panic attacks, suggesting a potential association between depression and panic attacks.

Although most respondents do not report either condition, panic attacks appear relatively more common among students who report depression.
